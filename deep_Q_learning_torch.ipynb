```python

import pickle
import json
import math
import scipy
import scipy.interpolate
import random
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import torch
from torch.cuda.amp import autocast
import copy
import concurrent.futures
def GameStateFactory(m,n,device):
    transition_map=torch.tensor([1,n,-n,-1], device=device)
    class GameState:
        def __init__(self,num_parallel_games=1,initial_state=None,num_shuffles=100):
            self.batch_size=num_parallel_games
            self.transition_map=transition_map
            self.state=None
            self.m=m
            self.n=n
            self.device=device
            if initial_state is None:
                L=np.zeros([self.batch_size,m*n],dtype=np.int32)
                for batch in range(self.batch_size):
                    L[batch,:self.m*self.n-1]=np.arange(1,self.m*self.n)
                self.state=torch.from_numpy(L.flatten()).to(self.device)
                self.scramble_state(num_shuffles)
            else:
                self.state=initial_state
        def scramble_state(self, num_shuffles):
            for random_move in range(num_shuffles):
                indices=torch.randint(0,len(self.transition_map),(self.batch_size,))
                actions=self.transition_map[indices]
                self.slide(actions, False)
        def slide(self,actions,stop_on_complete=True):
            places=(self.state == 0).nonzero()%(self.m*self.n)
            offsets=actions
            if stop_on_complete:
                mask=torch.logical_not(self.is_complete())
                offsets*=mask
            offsets=offsets.unsqueeze(1)
            negate=torch.logical_or(places+offsets<0,places+offsets>=self.m*self.n)
            offsets=(1-2*negate)*offsets
            indices=torch.arange(self.m*self.n*self.batch_size,device=self.device)
            mod_indices=(indices%(self.m*self.n)).reshape([-1,self.m*self.n])
            indices_diff=offsets*(mod_indices==places)-offsets*(mod_indices==places+offsets)
            self.state=self.state.gather(0,indices+indices_diff.flatten())
        def transition(self,indices):
            self.slide(self.transition_map.gather(0,indices.flatten()).flatten())
        def score(self):
            solved_state=torch.cat([torch.arange(1,self.m*self.n,device=self.device),torch.zeros(1,device=self.device)],0)
            state=self.state.reshape((-1,self.m*self.n))
            solved_state=solved_state.unsqueeze(0)
            return torch.sum(solved_state==state,1)
        def is_complete(self):
            return self.score().flatten()==self.m*self.n
        def partial_score(self):
            one=torch.ones(self.m*self.n*self.batch_size,device=self.device)
            zero=torch.zeros(self.m*self.n*self.batch_size,device=self.device)
            indices=torch.arange(self.batch_size*self.m*self.n,device=self.device)
            is_first_place=torch.logical_and(zero==indices, one==self.state)
            is_last_place=torch.logical_and((self.m*self.n-1)*one==indices,
                                            zero==self.state)
            first=torch.count_nonzero(is_first_place.reshape([-1,self.m*self.n]),1)
            last=torch.count_nonzero(is_last_place.reshape([-1,self.m*self.n]),1)
            return first*last
    return GameState
def groupoid_basis(p,q):
    amn=scipy.linalg.circulant(np.eye(3)[0])
    return scipy.linalg.block_diag(*[amn] * p * q)
    
class GroupoidDecompositionLayer(torch.nn.Module):
    def __init__(self,m,n,device='cuda'):
        super().__init__()
        self.device=device
        self.basis=torch.from_numpy(groupoid_basis(m,n).astype(np.float32)).to(device)
        self.m=m
        self.n=n
    def forward(self,tensor):
        components=torch.matmul(tensor, self.basis)
        return components.diagonal(offset=0, dim1=-1, dim2=-2).sum(-1)
  
class PermutationEquivariantLayer(torch.nn.Module):
    def __init__(self, output_dim, device):
        super().__init__()
        self.weight =torch.nn.Parameter(torch.empty([output_dim,2],requires_grad=True, device=device))
        self.bias=torch.nn.Parameter(torch.empty([output_dim], requires_grad=True, device=device))
        self.weight=torch.nn.init.uniform_(self.weight,-1e-8,1e-8)
        self.bias=torch.nn.init.uniform_(self.bias,-1e-8,1e-8)
        self.output_dim=output_dim
        self.myparameters = torch.nn.ParameterList([self.weight,self.bias])
    def forward(self,tensor):
        x1=torch.unsqueeze(torch.unsqueeze(tensor,2),3)*self.weight[:,1]
        x2=torch.transpose(torch.unsqueeze(torch.unsqueeze(tensor,2),3)*self.weight[:,0],2,1)
        return torch.sum(torch.sum(x1+x2,2),1)+self.bias.unsqueeze(0)

"""class ScaledDotProductAttention(torch.nn.Module):
    def __init__(self,
                 d_k,
                 device='cuda'):
        super().__init__()
        self.device = device
        self.scaler = np.sqrt(d_k)
    def forward(self, q, k, v, mask=None):
        '''
        # Argument
            q, k, v: (batch, sequence, out_features)
            mask:    (batch, sequence)
        '''
        score = torch.einsum('ijk,ilk->ijl', (q, k)) / self.scaler
        score = score - torch.max(score,
                                  dim=-1,
                                  keepdim=True)[0]  # softmax max trick

        score = torch.exp(score)
        if mask is not None:
            # suppose `mask` is a mask of source
            # in source-target-attention, source is `k` and `v`
            if len(mask.size()) == 2:
                mask = mask.unsqueeze(1).repeat(1, score.size(1), 1)
            # score = score * mask.float().to(self.device)
            score.data.masked_fill_(mask, 0)

        a = score / torch.sum(score, dim=-1, keepdim=True)
        c = torch.einsum('ijk,ikl->ijl', (a, v))

        return c"""

class GroupoidUnit(torch.nn.Module):
    def __init__(self,num_outputs,m,n, input_size, num_hidden_layers, num_hidden_units, hidden_unit_size, attn_scale, device):
            super().__init__()
            #self.attn=ScaledDotProductAttention(attn_scale,device=device)
            self.num_outputs=num_outputs
            self.m=m
            self.n=n
            self.input_size=input_size
            self.hidden_unit_size=hidden_unit_size
            self.num_hidden_units=num_hidden_units
            self.num_hidden_layers=num_hidden_layers
            self.groupoid=GroupoidDecompositionLayer(self.m,self.n,device)
            self.basis_dim=self.groupoid.basis.flatten().shape[0]
            self.basis_shape=list(self.groupoid.basis.shape)
            self.device=device
            layers=[]
            params=[]
            for output_unit in range(self.num_outputs):
                    for layer in range(self.num_hidden_layers):
                        for j in range(self.num_hidden_units):
                            layers.append(PermutationEquivariantLayer(self.hidden_unit_size,device))
                            params.append(layers[-1].weight)
                            params.append(layers[-1].bias)
                    layers.append(PermutationEquivariantLayer(self.basis_dim,device))
                    params.append(layers[-1].weight)
                    params.append(layers[-1].bias)
                    layers.append(torch.nn.BatchNorm1d(self.basis_dim,track_running_stats=False))
                    if device!='cpu':
                        layers[-1]=layers[-1].cuda()
                    params.append(layers[-1].weight)
                    params.append(layers[-1].bias)
            self.layers=layers
            self.params=params
            self.myparameters=torch.nn.ParameterList([p for l in self.layers for p in l.parameters()])
            self.sigmoid=torch.tanh
    def copy_from(self,unit):
        params=[]
        for i, layer in enumerate(self.layers):
            layer.weight=torch.nn.Parameter(unit.layers[i].weight.clone().to(self.device))
            layer.bias=torch.nn.Parameter(unit.layers[i].bias.clone().to(self.device))
            params.append(layer.weight)
            params.append(layer.bias)
        self.params=params
        self.myparameters=torch.nn.ParameterList([p for l in self.layers for p in l.parameters()])
    def forward(self, inputs):
                outputs=[]
                index=0
                for output_unit in range(self.num_outputs):
                    reshaped=torch.reshape(inputs,(-1,self.input_size,1))
                    #attention=self.attn(reshaped,reshaped,reshaped)
                    #attention=torch.reshape(attention,(-1,self.input_size))
                    #concat=torch.cat([inputs,attention],1)
                    concat=inputs
                    hidden_layers=[]
                    for layer in range(self.num_hidden_layers):
                        for j in range(self.num_hidden_units):
                            hidden_layers.append(torch.nn.LeakyReLU()(self.layers[index](concat)))
                            index+=1
                        concat=torch.cat(hidden_layers[-self.num_hidden_units*2:],1)
                    representation_layer=self.layers[index](concat)
                    index+=1
                    representation_layer=self.layers[index](representation_layer)
                    index+=1
                    representation_layer=torch.reshape(representation_layer,[-1]+self.basis_shape)
                    output=torch.reshape(self.groupoid(representation_layer),[-1,1])
                    outputs.append(output)
                return self.sigmoid(torch.nn.Dropout(p=0.25)(torch.cat(outputs,1))), self.sigmoid(torch.reshape(outputs[0],[-1,1]))
    def load_weights(weights):
        params=[]
        for i,layer in enumerate(self.layers):
            layer.weight=torch.nn.Parameter(weights[i*2].to(self.device))
            layer.bias=torch.nn.Parameter(weights[i*2+1].to(self.device))
            params.append(layer.weight)
            params.append(layer.bias)
        self.params=params
        self.myparameters=torch.nn.ParameterList([p for l in self.layers for p in l.parameters()])
        return len(self.params)
class Model(torch.nn.Module):
    def __init__(self, m,n,  num_groupoid_units, input_size, num_hidden_layers, num_hidden_units=10, hidden_unit_size=None, attn_scale=1, device='cuda'):
        super().__init__()
        if hidden_unit_size is None:
            hidden_unit_size=input_size
        self.units=[]
        self.params=[]
        self.device=device
        for i_unit in range(num_groupoid_units):
            self.units.append(GroupoidUnit(input_size,m,n,input_size, num_hidden_layers, num_hidden_units, hidden_unit_size, attn_scale,device))
            self.params+=self.units[-1].params
        self.parameters=torch.nn.ParameterList([p for unit in self.units for p in unit.parameters()])
    def copy_from(self,model):
        for i,unit in enumerate(self.units):
            unit.copy_from(model.units[i])
        self.parameters=torch.nn.ParameterList([p for unit in self.units for p in unit.parameters()])
        self.params=[p for unit in self.units for p in unit.params]
    def forward(self, inputs):
        inputs=inputs.type(torch.cuda.FloatTensor if self.device!='cpu' else torch.FloatTensor)
        for unit in self.units:
            inputs, out_first=unit(inputs)
        return out_first
    def weights_to_json(self):
        weights_all=[w.clone().detach().cpu().numpy().tolist() for w in self.params]
        return json.dumps(weights_all)
    def weights_from_json(self,s_weights):
        weights_all=json.loads(s_weights)
        weights_all=[torch.from_numpy(param).to(self.device) for w in self.params]
        for i,unit in enumerate(self.units):
            n=unit.load_weights(weights_all)
            weights_all=weights_all[n:]
        self.parameters=torch.nn.ParameterList([p for unit in self.units for p in unit.parameters()])
        self.params=[p for unit in self.units for p in unit.params]
class SlideAgent:
    def __init__(self, 
                 m, 
                 n, 
                 exploration_probability=0.9,
                 exploration_decay=0.07,
                 gamma=0.9997,
                 num_trials_per_update=5, 
                 max_turns_per_game=150,
                 max_cycle_penalty=0.1,
                 experience_replay_sample_size=64,
                 learning_rate=0.1,
                 num_hidden_layers=1,
                 num_hidden_units=1,
                 num_groupoid_units=3,
                 reward_time_penalty=0.99,
                 print_every_step=False,
                 device='cuda'):
        self.reward_time_penalty=reward_time_penalty
        self.num_trials_per_update=num_trials_per_update
        self.max_turns_per_game=max_turns_per_game
        self.max_cycle_penalty=max_cycle_penalty
        self.print_every_step=print_every_step
        self.experience_replay_sample_size=experience_replay_sample_size
        self.m=m
        self.n=n
        self.gamma=gamma
        self.exploration_probability=exploration_probability
        self.exploration_decay=exploration_decay
        self.num_actions=4
        self.GameState=GameStateFactory(m,n,device)
        self.input_size=self.m*self.n#*2        
        self.experience_replay_buffer=torch.zeros([0,self.input_size],dtype=torch.int32, device=device)
        self.dqn=Model(m,n,num_groupoid_units, self.input_size, num_hidden_layers, num_hidden_units, device=device)
        self.dqn_target=Model(m,n,num_groupoid_units, self.input_size, num_hidden_layers, num_hidden_units, device=device)
        self.opt=torch.optim.Adam(self.dqn.parameters(), lr=learning_rate)
        self.update_target()
        self.device=device
        self.num_turns=0
        self.num_turns0=0
        self.lr_scheduler=torch.optim.lr_scheduler.ReduceLROnPlateau(self.opt)
    def update_target(self):
        self.dqn_target.copy_from(self.dqn)
    def all_action_Q(self,state,target=False):
        q_values=[]
        state1=torch.reshape(state,[-1,1,self.m*self.n]).repeat([1,self.num_actions,1])
        actions=torch.arange(self.num_actions,device=self.device).unsqueeze(0).unsqueeze(-1).repeat([state.shape[0],1,1])
        q_values=self.predict_q(state1,actions,target)
        return q_values
    def choose_action(self,state,target=False):
        q_values=self.all_action_Q(state,target=target)
        return torch.max(q_values.reshape([-1,self.num_actions]),dim=-1)
    def predict_q(self,state,action, target=False):#value function when we only use the current state
        model_input_vector,rewards=self.encode_input(state,action)
        model_input_vector=model_input_vector.reshape([-1,self.input_size])
        if target:
            with torch.no_grad():
                return self.dqn_target(model_input_vector).flatten()*self.gamma+rewards
        else:
            return self.dqn(model_input_vector).flatten()*self.gamma+rewards
    def encode_input(self,state,action):
        g=self.GameState(np.prod(state.shape[:-1]),initial_state=state.flatten())
        r1=self.reward(g)
        g.transition(action.flatten())
        state2=g.state.reshape(state.shape)
        r2=self.reward(g)
        #S=torch.cat([state,state2],len(state.shape)-1)
        S=state2
        return S, r2-r1+self.max_cycle_penalty*(1.0-2.0*torch.logical_not(g.is_complete())) #not advancing in reward after many turns should trigger the cycle penalty.
    def state_reward(self,state2):
        g=self.GameState(np.prod(state2.shape[:-1]), state2.flatten())
        return self.reward(g)
    def reward(self,game): #cumulative reward associated with the current state, a greedy algorithm would just use this as a value function
                            #we can take the difference over successive states to get the instantaneous reward which is used in the Q update
        score_adjusted= 0        
        score_adjusted=2.0*(game.is_complete())
        score_adjusted+=0.5*game.partial_score()*torch.exp(2.0*game.score()/self.m*self.n)
        score_adjusted-=torch.exp(0.5*(1.0-game.score()/self.m*self.n))
        #score_adjusted*=self.reward_time_penalty**(len(replay))
        half=0.5*torch.ones(game.batch_size,device=self.device)
        return torch.minimum(half,torch.maximum(-half,score_adjusted/2.0))
    def play_game(self,num_shuffles, batch_size=1, use_target=False):
        replay=[]
        game=self.GameState(batch_size,None,num_shuffles)
        while torch.any(game.is_complete()).cpu().numpy():
            game=self.GameState(batch_size,None,num_shuffles) #just in case
        max_iterations=self.max_turns_per_game+random.randrange(0,5)
        for iteration in range(max_iterations):
            if np.random.rand()<self.exploration_probability:
                action=torch.randint(0,self.num_actions,(batch_size,), device=self.device)
            else:
                with torch.no_grad():
                    action=self.choose_action(game.state.reshape([-1,self.m*self.n]),target=use_target)[1]
            game.transition(action)
            is_last_turn=(max_iterations<=iteration+1)
            replay.append([game.state])
            if self.print_every_step:
                print(f'{np.reshape(game.state.cpu().numpy(),[-1,self.m,self.n])}'.replace(' 0',' _').replace('[0','[_'))
            if torch.all(game.is_complete()):
                break
        if self.print_every_step:
            print(replay[-1][-1])
        return replay,game.is_complete(), game.score()
    
    def update_weights(self,inputs_device):
        Q1=self.choose_action(inputs_device,target=True)[0]
        V0=self.dqn(inputs_device)
        loss=torch.nn.L1Loss(reduction='mean')(V0.flatten(),Q1.flatten())
        self.opt.zero_grad()
        loss.backward()
        self.opt.step()
        self.lr_scheduler.step(loss)
        return loss.detach().cpu().numpy()
            
    def train_once(self,epoch):
        transition_batch=[]
        torch.zeros([0,self.m*self.n],device=self.device)
        scores=[]
        win_rate=0
        difficulty=min(self.max_turns_per_game,max(int(0.15*epoch),int(np.sqrt(epoch)*0.5+2)))
        difficulty0=min(self.max_turns_per_game,max(int(0.15*epoch),int(np.sqrt(epoch-1)*0.5+2)))
        if difficulty0<difficulty:
            self.exploration_probability=max(self.exploration_probability,0.25)
        if self.experience_replay_buffer.shape[0]>5000*self.experience_replay_sample_size:
            self.experience_replay_buffer=self.experience_replay_buffer[1:,:]
            self.num_turns0=self.num_turns
        results=self.play_game(difficulty,self.num_trials_per_update)
        replay_buffer, won, score=results
        for move in replay_buffer:
            self.experience_replay_buffer=torch.cat([self.experience_replay_buffer,move[0].reshape([-1,self.m*self.n])],0)
        scores+=(score.tolist())
        win_rate=np.sum(won.cpu().numpy())/self.num_trials_per_update
        self.num_turns+=self.num_trials_per_update*self.max_turns_per_game
        loss=None
        if self.num_turns-self.num_turns0>=self.experience_replay_sample_size:   
            perm = torch.randperm(self.experience_replay_buffer.shape[0])
            batch_indices = perm[:self.experience_replay_sample_size]
            inputs=self.experience_replay_buffer[batch_indices]
            loss=self.update_weights(inputs)
        self.exploration_probability=(1-self.exploration_decay)*self.exploration_probability
        mean_score=np.mean(scores)
        all_rewards=self.state_reward(self.experience_replay_buffer[-5000*self.num_trials_per_update:])
        all_rewards=all_rewards.reshape([-1,self.num_trials_per_update]).transpose(0,1)
        return win_rate, all_rewards.cpu().numpy(), loss, mean_score, difficulty
        
agent=SlideAgent(4,4,print_every_step=False, device='cuda')
print(sum([np.prod(w.shape) for w in agent.dqn.params]))
iteration=0
losses=[]
win_rates=[]
while True:
    win_rate, rewards, loss, mean_score, difficulty=agent.train_once(iteration+1)
    win_rates.append(win_rate)
    iteration+=1
    losses.append(loss)
    if iteration%1==0:
        print(f'win rate:{win_rate} on iteration {iteration}, loss: {loss}, mean score: {mean_score}, difficulty: {difficulty}, eps: {agent.exploration_probability}')
    if iteration%5==0:
        agent.update_target()
    if iteration%25==0:
        plt.subplot(1,2,1)
        plt.plot(rewards[-100:].T)
        plt.subplot(1,2,2)
        plt.scatter(np.arange(len(losses)), losses)
        plt.scatter(np.arange(len(win_rates)),win_rates)
        plt.show()
    if iteration%50==0:
        open('dqn_weights.json', 'w+').write(agent.dqn.weights_to_json())
        open('dqn_target_weights.json', 'w+').write(agent.dqn_target.weights_to_json())
        np.savetxt('rewards.np', rewards)
        np.savetxt(f'training_loss.np', np.array(losses))
        np.savetxt(f'win_rate.np', np.array(win_rates))
```
